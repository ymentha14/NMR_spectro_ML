<a id='ensemble_method'></a>
## 5. Ensemble methods

<div class="alert alert-block alert-success">
In order to take advantage of the 3 datasets, 3 similar NN were independently trained on the 3 dataets (3,5 and 7${\\A}$). Then, each of the output was assembled thanks to the following formula, which privileges similar results, and does not take outliers into account
</div>

$\bar{y} = \frac{\sum_{j \in S} \sum_{j \in S,i \neq j}\frac{1}{a_{ij}}y_{i}}{\sum_{j \in S}\sum_{i \in S,i \neq j}\frac{1}{a_{ij}}},a_{ij} = |y_{i} - y_{j}|$

<div class="alert alert-block alert-success">
The following results were obtained performing a RFE keeping 2160 features on 20000 samples using a 5Fold CV.
</div>

```python
with open('../pickle_files/scoresbigbibo2.pickle', 'rb') as f: 
    scores = pickle.load(f)
```

```python
tab = [ ["", 'MSE', 'MAE'],
        ["Mean method", '{:0.2f} +/- {:0.4f}'.format(np.mean(scores['mean_method']['mse']), np.std(scores['mean_method']['mse'])), 
                  '{:0.2f} +/- {:0.4f}'.format(np.mean(scores['mean_method']['mae']), np.std(scores['mean_method']['mae']))],
       ["Custom Method", '{:0.2f} +/- {:0.4f}'.format(np.mean(scores['custom_method']['mse']), np.std(scores['custom_method']['mse'])), 
                  '{:0.2f} +/- {:0.4f}'.format(np.mean(scores['custom_method']['mae']), np.std(scores['custom_method']['mae']))]
      ]

print(tabulate(tab, tablefmt='fancy_grid'))
```

<div class="alert alert-block alert-success">
Overall, we see that the custom method described above performs roughly as the arithmetic mean.
</div>
